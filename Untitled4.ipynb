```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('Json.json','r') as my:
    dt = my.read()
mayabeque = json.loads(dt)

```

```python
def llenar_lista(filtro1:str,filtro2:str,lista1:list)->list:
    lista = []
    for keys,values in mayabeque.items():
        if keys == filtro1:
            for clave,valor in values.items():
                 if clave in lista1:
                     for key,value in valor.items():
                        if key == filtro2:
                            lista.extend(value)
    return lista
```

```python
# creando matrices del saldo medio mensual de las provincias occidentales( ao 2011-2022 )

smmPin = np.array(mayabeque["Provinciassmm"]["Pinar-del-Rio"]["smm-11-22"]) # Pinar del Rio
smmArt = np.array(mayabeque["Provinciassmm"]["Artemisa"]["smm-11-22"]) # Artemisa
smmMay = np.array(mayabeque["Provinciassmm"]["Mayabeque"]["smm-11-22"]) # Mayabeque
smmHab = np.array(mayabeque["Provinciassmm"]["La-Habana"]["smm-11-22"]) # La Habana

# matrices del saldo medio mensual del resto del pais( ao 2011-2022 )
smmMat = np.array(mayabeque["Provinciassmm"]["Matanzas"]["smm-11-22"]) # Matanzas
smmVill = np.array(mayabeque["Provinciassmm"]["Villa-Clara"]["smm-11-22"]) # Villa Clara
smmCien = np.array(mayabeque["Provinciassmm"]["Cienfuegos"]["smm-11-22"]) # Cienfuegos
smmSanc = np.array(mayabeque["Provinciassmm"]["Sancti-Spiritus"]["smm-11-22"]) # Sancti Spiritus
smmCieg = np.array(mayabeque["Provinciassmm"]["Ciego-de-Avila"]["smm-11-22"]) # Ciego de Avila
smmCam = np.array(mayabeque["Provinciassmm"]["Camaguey"]["smm-11-22"]) # Camaguey
smmTun = np.array(mayabeque["Provinciassmm"]["Las-Tunas"]["smm-11-22"]) # Las Tunas
smmHol = np.array(mayabeque["Provinciassmm"]["Holguin"]["smm-11-22"]) # Holguin
smmGran = np.array(mayabeque["Provinciassmm"]["Granma"]["smm-11-22"]) # Granma
smmSant = np.array(mayabeque["Provinciassmm"]["Santiago-de-Cuba"]["smm-11-22"]) # Santiago de Cuba
smmGuan = np.array(mayabeque["Provinciassmm"]["Guantanamo"]["smm-11-22"]) # Guantanamo
smmIsla = np.array(mayabeque["Provinciassmm"]["Isla-de-la-Juventud"]["smm-11-22"]) # Isla de la Juventud

# promedios de los saldos medios mensuales de las provincias occidentales
proPin = smmPin.mean().round()
proArt = smmArt.mean().round()
proMay = smmMay.mean().round()
proHab = smmHab.mean().round()

# promedios de los saldos medios mensuales del resto del pais

proMat = smmMat.mean().round()
proVill =smmVill.mean().round()
proCien =smmCien.mean().round()
proSanc =smmSanc.mean().round()
proCieg =smmCieg.mean().round()
proCam =smmCam.mean().round()
proTun =smmTun.mean().round()
proHol =smmHol.mean().round()
proGran =smmGran.mean().round()
proSant =smmSant.mean().round()
proGuan =smmGuan.mean().round()
proIsla =smmIsla.mean().round()
 
# promedio del saldo medio mensual de Cuba( ao 2011-2022 )

smmCuba = (proPin + proArt + proMay + proHab + proMat + proVill + proCien + proSanc 
           + proCieg + proCam + proTun + proHol + proGran + proSant + proGuan + proIsla )/16

# redondeando el promedio del saldo medio mensual de Cuba
smmCuba = smmCuba.round()
 
#creando un grafico de lineas con las provincias occidentales
plt.figure(figsize = [9,7])
plt.plot([x for x in mayabeque['annos'] if x >= 2011], smmPin, marker = 'o', linestyle = '-', color = 'g', label = 'Pinar del Rio')
plt.plot([x for x in mayabeque['annos'] if x >= 2011], smmArt, marker = 'o', linestyle = '-', color = 'm', label = 'Artmisa')
plt.plot([x for x in mayabeque['annos'] if x >= 2011], smmMay, marker = 'd', linestyle = '-', color = 'b', label = 'Mayabeque')
plt.plot([x for x in mayabeque['annos'] if x >= 2011], smmHab, marker = 'o', linestyle = '-', color = 'r', label = 'La Habana')
plt.xlabel("Aos")
plt.ylabel("Saldo medio mensual")
plt.title("Saldo medio mensual de las provincias occidentales del pas")
plt.xticks([2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022])
plt.yticks([500,800,1000,1200,1500,2000,2500,3000,3500,4000,4500])
plt.axhline(y = smmCuba,color = 'k', linestyle = '--', linewidth = 4, label = f"media de Cuba: {smmCuba}")
plt.legend()
plt.show()
```
